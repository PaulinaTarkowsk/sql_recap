```python
import pandas as pd

# Załaduj dane z pliku CSV do DataFrame
imdb_reviews = pd.read_csv('https://raw.githubusercontent.com/marcin119a/data/refs/heads/main/imdb_reviews_lemmatized.csv')
```

```python
imdb_reviews.head(15)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(max_features = 1000)
```

```python
review_vectors = vectorizer.fit_transform(imdb_reviews["review_lemmatized"])
```

```python
X = review_vectors.toarray()
print(X)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, imdb_reviews['sentiment'], test_size=0.3, random_state=42)

model = LogisticRegression(max_iter = 1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print("Raport z metrykami modelu:")
print(classification_report(y_test, y_pred))
```
