```python
from dash import html, dcc, Input, Output

df_bpm = df_bpm.sort_values(['top year','bpm'], ascending=[False, False])


# create app
app = dash.Dash()

min=df_bpm['top year'].min()
max=df_bpm['top year'].max()


# configure app layout
app.layout = html.Div([html.H1('Spotify bpm'),
        html.Div(children=[html.Label('Year:', style={'font-weight': 'bold', "text-align": "center"}),
        dcc.Slider(id='year',
        min=min,
        max=max,
        step=1,
        marks={i: '{}'.format(i) for i in range(1,2021,1)},
        value=2017
        ),
        dcc.Graph(id='graph',style={'height': '600px'})
    ])])



# Define callback to update graph
@app.callback(
    [Output('graph', 'figure')],
    [Input("year", "value")]
)
def update_figure(year):
    df_bpm1 = df_bpm[df_bpm['top year'] == year]
    chart = px.bar(
        df_bpm1.head(15),
        x="title",
        y="bpm",
        title='Top 15 titles by bpm')
    return chart


thread = threading.Thread(target=app.run_server)
thread.start()
output.serve_kernel_port_as_iframe(port=8050, height=900)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import dash
from dash import html, dcc, Input, Output, State
import threading
from google.colab import output, drive

drive.mount('/content/drive')
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Spotify_top100/spotify_top100.csv')


year_list = list(df['top year'].drop_duplicates())

metric_dict = {'bpm': 'bpm',
               'nrgy': 'nrgy',
               'dB': 'dB',
               'live': 'live',
               'val': 'val'}

# create app
app = dash.Dash()

# configure app layout
app.layout = html.Div([html.H1('Spotify'),
        html.Div(children=[html.Label('Year:', style={'font-weight': 'bold', "text-align": "center"}),
        dcc.Dropdown(
        id='top_year',
        options=[{'value': (top_year), 'label': top_year}
                 for top_year in year_list],
        value=year_list[0]
        ),html.Label('Metric:', style={'font-weight': 'bold', "text-align": "center"}),
        dcc.Dropdown(
        id='metric',
        options=[{'value': metric, 'label': metric_label}
                 for metric, metric_label in metric_dict.items()],
        value='bpm'
       ),
        dcc.Graph(id='graph',style={'height': '500px'})
    ])])



# Define callback to update graph
@app.callback(
    Output('graph', 'figure'),
    [State('metric', 'value'),
     Input('top_year', 'value')]
)

def update_figure(metric,top_year):
    df_bpm1 = df_bpm[df_bpm['top year'] == top_year].sort_values([metric], ascending=False)
    chart = px.bar(
        df_bpm1.head(15),
        x="title",
        y=df_bpm1[metric].head(15),
        title='Top 15 titles by '+metric)
    return chart



thread = threading.Thread(target=app.run_server)
thread.start()
output.serve_kernel_port_as_iframe(port=8050, height=900)
```
