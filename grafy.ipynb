```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
plt.axis("off")
```

```python
G = nx.Graph()
G.add_edge('B', 'A')
#pos={'B': (0, 0), 'A': (1, 3), 'C': (2, 1), 'D': (4, 2), 'E': (4, 0)}
nx.draw_networkx(G, pos=None, with_labels=True)
plt.show()
```

```python
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'),('C', 'D'),('D', 'E')]
G = nx.DiGraph()
G.add_edges_from(edge_list)
nx.draw_networkx(G, pos=None, with_labels=True)
plt.show()
```

```python
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'),('C', 'D'),('D', 'E')]
G = nx.DiGraph()
G.add_edges_from(edge_list)
pos={'B': (0, 0), 'A': (1, 3), 'C': (2, 1), 'D': (4, 2), 'E': (4, 0)}
nx.draw_networkx(G,
    font_size = 24,
    node_size = 1000,
    node_color = "blue",
    edgecolors = "black",
    linewidths = 2,
    width = 2,
)
```

```python
width = [3,3,3,1,1,1]
nx.draw_networkx(
    G,
    width=width
)
```

```python
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'), ('C', 'D'), ('D', 'E')]
G = nx.Graph()
G.add_edges_from(edge_list)
nx.draw_networkx(G, pos=None, with_labels=True)
plt.show()
```

```python
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos=pos, with_labels=True)
plt.show()
```

```python
pos = nx.circular_layout(G)
nx.draw_networkx(G, pos=pos, with_labels=True)
plt.show()
```

```python
pos = nx.random_layout(G)
nx.draw_networkx(G, pos=pos, with_labels=True)
plt.show()
```

```python
ax=plt.gca()
ax.margins(1)
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'), ('C', 'D'), ('D', 'E')]
pos={'B': (0, 0), 'A': (1, 3), 'C': (2, 1), 'D': (4, 2), 'E': (4, 0)}
G = nx.Graph()
G.add_edges_from(edge_list)
nx.draw_networkx(G, pos=pos, with_labels=True)
plt.show()
```

```python
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'),('C', 'D'),('D', 'E')]
node_color = ['red','blue','red','red','blue']
edge_width = [3,3,3,1,1,1]
G = nx.DiGraph()
G.add_edges_from(edge_list)
nx.draw_networkx(G, pos=None, with_labels=True, node_color = node_color, width = edge_width)
plt.show()
```
