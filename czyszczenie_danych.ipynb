```python
from sklearn.impute import SimpleImputer
import pandas as pd

# Tworzenie obiektu SimpleImputer z uyciem strategii 'median'
imputer = SimpleImputer(strategy="median")

# Usunicie kolumny 'minhr' przed imputacj
df_without_minhr = df.drop("minhr", axis=1)

# Wyfiltrowanie kolumn liczbowych
numeric_cols = df_without_minhr.select_dtypes(include=["number"]).columns

# Dopasowanie i transformacja (uzupenienie median wartoci brakujcych)
df_numeric_imputed = pd.DataFrame(imputer.fit_transform(df_without_minhr[numeric_cols]), columns=numeric_cols)

# Zastpienie starych kolumn liczbowych nowymi w DataFrame bez kolumny 'minhr'
df_without_minhr[numeric_cols] = df_numeric_imputed

# Jeli chcesz doda kolumn 'minhr' z powrotem (lub moesz j pomin, jeli nie jest potrzebna)
#df_without_minhr["minhr"] = df["minhr"]

df = df_without_minhr

# Sprawdzenie, czy wartoci zostay uzupenione
print(df.isnull().sum())

```

```python
df["chestpain"] = df["chestpain"].replace("nontypiccal", "nontypical")
df["chestpain"] = df["chestpain"].replace("nontypica", "nontypical")
df["chestpain"] = df["chestpain"].replace("asymptomaticc", "asymptomatic")

df["chestpain"].value_counts()
```

```python
# How many unique values are in thal variable

nunique_thal = df["thal"].nunique()
print(f"Number of unique values in thal variable: {nunique_thal}")

print("--------------------------------")

# How many observations (rows) are per every value of thal variable

df["thal"].value_counts()
```

```python
df["thal"] = df["thal"].str.lower()
df["thal"] = df["thal"].replace("reversabble", "reversable")
df["thal"].value_counts()
```

```python
# How many unique values are in ahd variable

nunique_ahd = df["ahd"].nunique()
print(f"Number of unique values in ahd variable: {nunique_ahd}")

print("--------------------------------")

# How many observations (rows) are per every value of ahd variable

df["ahd"].value_counts()
```

```python
df["ahd"] = df["ahd"].str.capitalize()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

dataset_path = '/drive/My Drive/Machine Learning/Google Colab/heart_data_clean.csv'
df = pd.read_csv(dataset_path)

# Inicjalizacja OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Przeksztacenie zmiennych kategorycznych
encoded_data = encoder.fit_transform(df[["sex", "chestpain", "exang", "thal", "fbs", "ahd"]])

# Tworzenie DataFrame z zakodowanymi danymi
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(["sex", "chestpain", "exang", "thal", "fbs", "ahd"]))

# czenie zakodowanych danych z oryginalnym DataFrame
df_encode = pd.concat([df, encoded_df], axis=1)

# Usunicie oryginalnych kolumn kategorycznych
df_encode.drop(columns=["sex","chestpain", "exang", "thal", "fbs", "ahd"], inplace=True)

# Wywietlenie kocowego DataFrame
print(df_encode.tail(10))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

dataset_path = '/drive/My Drive/Machine Learning/Google Colab/heart_data_numerics.csv'
df = pd.read_csv(dataset_path)


# Inicjalizacja StandardScaler
scaler = StandardScaler()

# Standaryzacja danych
heart_data_standardized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Wywietlenie standaryzowanych danych
print(heart_data_standardized)

heart_data_standardized.to_csv("/drive/My Drive/Machine Learning/Google Colab/heart_data_standardized.csv", index=False)
```

```python
X = heart_data_standardized.drop("ahd_Yes", axis=1)
y = heart_data_standardized["ahd_Yes"]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Split train data into valid and test set (70% training set, 30% valid set). In result we have: 70% training set, 15% valid set, 15% test set
X_test, X_valid, y_test, y_valid = train_test_split(X_test, y_test, test_size=0.5, random_state=42)


print(X_train.shape, X_valid.shape, X_test.shape)
```
