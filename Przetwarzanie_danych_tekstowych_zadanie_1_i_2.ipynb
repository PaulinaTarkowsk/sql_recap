Napisz funkcję, która jako argument przyjmie tekst recenzji i zwróci tekst w oczyszczonej formie:

Wszystkie słowa będą składać się z małych liter,

W tekście pozostaną tylko i wyłącznie litery (Wskazówka: użyj wyrażenia regularnego),

Tekst zostanie wyczyszczony ze słów przystankowych.
Możesz wykorzystać dowolną z poznanych opcji na ich usunięcie.

```python
def clean_review(text):
    # Przekształć tekst do małych liter
    text = text.lower()
    # Usuń wszystko co nie jest literą
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenizuj tekst
    words = word_tokenize(text)
    # Połącz słowa z powrotem w string
    return ' '.join(words)
```

```python
# Dodanie nowej kolumny review_cleaned poprzez zastosowanie funkcji clean_review
df['review_cleaned'] = df['review_text'].apply(clean_review)
```

```python
def lemmatize_text(text):
    doc = nlp(text)
    lemmatized_words = [token.lemma_ for token in doc]

    lemmatized_sentence = " ".join(lemmatized_words)

    return lemmatized_sentence
```
